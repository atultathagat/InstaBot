## 1. Login to Instagram Handle using username and password

```python
def login_to_instagram():
    driver.get('https://www.instagram.com/')
    username_field=wait.until(EC.presence_of_element_located((By.NAME, 'username')))
    password_field=wait.until(EC.presence_of_element_located((By.NAME, 'password')))
    username_field.send_keys(username)
    password_field.send_keys(password)
    login_btn=driver.find_elements_by_tag_name('button')
    login_btn[1].click()
```

```python
# 1.1 logging to my instagram handle
login_to_instagram()
```

## 2. Typing “food” in search bar and printing all the names of the Instagram Handles that are displayed in list after typing “food”

```python
def print_name_of_instagram_handles():
    driver.get('https://www.instagram.com/')
    search_text=wait.until(EC.presence_of_element_located((By.XPATH, '//input[contains(@class,"XTCLo")]')))
    search_text.clear()
    search_text.send_keys('food')
    foods=wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'fuqBx')))
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'fuqBx')))
    food_handles=driver.find_elements_by_xpath('//span[@class="Ap253"]')
    for i in food_handles:
        if not i.text.startswith('#'):
            print(i.text)
```

```python
# 2.1 printing all the names of the Instagram Handles that are displayed in list after typing “food” in the search bar
print_name_of_instagram_handles()
```

## 4. Following and Unfollowing given instagram handle “So Delhi” by searching and then opening it

```python
def follow():
    follow_btn=wait.until(EC.presence_of_element_located((By.XPATH, '//button[contains(@class,"_5f5mN")]')))
    if follow_btn.text=='Follow':
        follow_btn.click()
    else:
        print('Already following')

def unfollow():
    btn= wait.until(EC.presence_of_element_located((By.XPATH, '//button[contains(@class,"_5f5mN")]')))
    if btn.text=='':
        btn.click()
        unfollow_btn=wait.until(EC.presence_of_element_located((By.XPATH, '//button[contains(@class,"aOOlW")]')))
        unfollow_btn.click()
    else:
        print('Already unfollowed')
```

## 5. Liking the top 30 posts of the ‘dilsefoodie' and then disliking it

```python
def like_30_posts():
    driver.get('https://www.instagram.com/')
    search_text=wait.until(EC.presence_of_element_located((By.XPATH, '//input[contains(@class,"XTCLo")]')))
    search_text.send_keys('dilsefoodie')
    dilsefoodie=wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'yCE8d')))
    dilsefoodie.click()
    post=wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"v1Nh3")]')))
    post.click()
    next_btn=wait.until(EC.presence_of_element_located((By.XPATH,'//a[contains(@class," _65Bje")]')))
    liked=0
    while True:
        like_btn= wait.until(EC.presence_of_element_located((By.XPATH, 
                                                             '//section[contains(@class,"Slqrh")]/span/button/div/span')))
        data=BeautifulSoup(like_btn.get_attribute('innerHTML'),'html.parser')
        like=data.svg.attrs['aria-label']
        if like=='Like':
            like_btn.click()
            print(liked,'Post Liked')
        else:
            print(liked,'Post already Liked')
        if liked==30:
            break
        liked+=1    
        next_btn.click()    

def unlike_30_posts():
    driver.get('https://www.instagram.com/')
    search_text=wait.until(EC.presence_of_element_located((By.XPATH, '//input[contains(@class,"XTCLo")]')))
    search_text.send_keys('dilsefoodie')
    dilsefoodie=wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'yCE8d')))
    dilsefoodie.click()
    post=wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"v1Nh3")]')))
    post.click()
    next_btn=wait.until(EC.presence_of_element_located((By.XPATH,'//a[contains(@class," _65Bje")]')))
    unliked=0
    while True:
        unlike_btn= wait.until(EC.presence_of_element_located((By.XPATH, 
                                                             '//section[contains(@class,"Slqrh")]/span/button/div/span')))
        data=BeautifulSoup(unlike_btn.get_attribute('innerHTML'),'html.parser')
        unlike=data.svg.attrs['aria-label']
        if unlike=='Unlike':
            unlike_btn.click()
            print(unliked,'Post Unliked')
        else:
            print(unliked,'Post already Unliked')
        if unliked==30:
            break
        unliked+=1    
        next_btn.click()    
```

```python
# 5.1 Liking the top 30 posts of the ‘dilsefoodie'
like_30_posts()
```

```python
# 5.2 Unliking the top 30 posts of the ‘dilsefoodie'
unlike_30_posts()
```

## 6.1 Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’

```python
def get_usernames(username,follow_flag=True):
    driver.get('https://www.instagram.com/'+username)
    if follow_flag:
        btn=wait.until(EC.presence_of_element_located((By.XPATH, '//ul[@class="k9GMp "]/li[2]/a')))
    else:
        btn=wait.until(EC.presence_of_element_located((By.XPATH, '//ul[@class="k9GMp "]/li[3]/a')))
    btn.click()
    wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="isgrP"]')))
    current_height=driver.execute_script('return document.querySelector(".isgrP").scrollHeight;')      
    while True:
        driver.execute_script('document.querySelector(".isgrP").scrollTo(0,arguments[0]);',current_height)
        time.sleep(1)
        usernames=driver.find_elements_by_xpath('//a[contains(@class,"FPmhX notranslate  _0imsa ")]')
        if len(usernames)>500:
            break;  
        new_height=driver.execute_script('return document.querySelector(".isgrP").scrollHeight;') 
        if new_height==current_height:
            break
        current_height=new_height     
    usernames_set=set()
    for username in usernames:
        data=BeautifulSoup(username.get_attribute('outerHTML'),'html.parser')
        usernames_set.add(data.get_text())
    return usernames_set
```

```python
# Extracting usernames of the first 500 followers of ‘foodtalkindia’
foodtalkindia_followers=get_usernames('foodtalkindia')
```

```python
# Printing usernames of the first 500 followers of ‘foodtalkindia’
for follower in list(foodtalkindia_followers)[:500]:
    print(follower)
```

```python
# Extracting usernames of the first 500 followers of ‘sodelhi’
sodelhi_followers=get_usernames('sodelhi')
```

```python
# Printing usernames of the first 500 followers of ‘sodelhi’
for follower in list(sodelhi_followers)[:500]:
    print(follower)
```

## 7. Check the story of ‘coding.ninjas’ with given conditions (story already seen, user has no story or view the story if not seen yet)

```python
def check_story(username):
    driver.get('https://www.instagram.com/'+username)
    try:
        story=wait.until(EC.presence_of_element_located((By.XPATH, '//header[@class="vtbgv "]//canvas')))
        if story.get_attribute('height')=='166':
            print("Already seen the story")
        else:
            story_btn=wait.until(EC.presence_of_element_located((By.XPATH, '//header[@class="vtbgv "]//span/img')))
            story_btn.click()
    except TimeoutException:
        print("The user has no story")
```

```python
check_story('coding.ninjas')
```
